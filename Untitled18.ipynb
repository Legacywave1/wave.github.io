```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

import plotly
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
df.rename(columns= {'Province/State': 'Province', 'Country/Region' : 'Country'}, inplace = True)

death_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv')
death_df.rename(columns= {'Province/State': 'Province', 'Country/Region' : 'Country'}, inplace = True)

recover_df = pd.read_csv('https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv')
recover_df.rename(columns= {'Province/State': 'Province', 'Country/Region' : 'Country'}, inplace = True)
```

```python
df['Status'] = 'Confirmed'
death_df['Status'] = 'Death'
recover_df['Status'] = 'Recovered'
confirmed_df = df.melt(id_vars = ['Province', 'Country', 'Lat', 'Long', 'Status'], value_name = 'Count', var_name = 'Date')
death_df = death_df.melt(id_vars = ['Province', 'Country', 'Lat', 'Long', 'Status'], value_name = 'Count', var_name = 'Date')
recover_df = recover_df.melt(id_vars = ['Province', 'Country', 'Lat', 'Long', 'Status'], value_name = 'Count', var_name = 'Date')
```

```python
total_data = pd.concat([confirmed_df, death_df, recover_df])
```

# Building KPI using plotly


```python
fig = go.Figure()

fig.add_trace(go.Indicator(
    mode = 'number+delta',
    value = TotalConfirmed['Count'].sum(),
    delta = {'reference': TotalConfirmed_MinusOne['Count'].sum(), 'relative': True},
    title = {'text': 'Confirmed'},
    domain = {'row' : 0, 'column' : 0}
))

fig.add_trace(go.Indicator(
    mode = 'number+delta',
    value = TotalRecovered['Count'].sum(),
    delta = {'reference': TotalRecovered_MinusOne['Count'].sum(), 'relative': True},
    title = {'text': 'Recovered'},
    domain = {'row' : 0, 'column' : 1}
))

fig.add_trace(go.Indicator(
    mode = 'number+delta',
    value = TotalDeath['Count'].sum(),
    delta = {'reference': TotalDeath_MinusOne['Count'].sum(), 'relative': True},
    title = {'text': 'Death'},
    domain = {'row' : 0, 'column' : 2}
))

fig.update_layout(grid = {'rows': 1, 'columns': 3, 'pattern': 'independent'},
                  template = {'data' : {'indicator' : [{
                      'title' : {'text' : 'Speed'},
                      'mode' : 'number+delta+gauge',
                      'delta' : {'reference' : 90}
                  }]}})
fig.show()
```

```python
def RecentCases_Top10Countries(Status):
    Top10Countries = total_data[(total_data['Status']==Status) & (total_data.index==max(total_data[total_data['Status']==Status].index))].sort_values(by='Count',ascending = False).head(10)[['Country','Count']]
    return Top10Countries
```

```python
Top10CountriesRecovered = RecentCases_Top10Countries('Recovered')
Top10CountriesConfirmed = RecentCases_Top10Countries('Confirmed')
Top10CountriesDeath = RecentCases_Top10Countries('Death')
```

```python
Top10CountriesConfirmed
```

```python
def Display_Top10Country(Top10Countries,Status):
    Top10Countries = Top10Countries.reset_index()
    Top10Countries = Top10Countries[['Country','Count']]
    date =max(total_data[total_data['Status']==Status].index)
    date = date.date()
    print(Status +' cases till:',date)
    return Top10Countries
```

```python
Top10CountriesRecovered = Display_Top10Country(Top10CountriesRecovered,'Recovered')
Top10CountriesRecovered.style.background_gradient(cmap = 'Reds')
```

```python
Top10CountriesConfirmed = Display_Top10Country(Top10CountriesConfirmed,'Confirmed')
Top10CountriesConfirmed.style.background_gradient(cmap = 'Reds')
```

```python
Top10CountriesDeath = Display_Top10Country(Top10CountriesConfirmed,'Death')
Top10CountriesDeath.style.background_gradient(cmap = 'Reds')
```

```python
date =max(total_data[total_data['Status']=='Confirmed'].index)
date = date.date()
px.bar(Top10CountriesConfirmed,x='Country',y='Count',color='Country',title='Top 10 Recovered cases till '+str(date))
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

fig = make_subplots(rows=3, cols=1)

fig.add_trace(
    go.Bar(x=Top10CountriesRecovered['Country'],y=Top10CountriesRecovered['Count'],text = Top10CountriesRecovered['Count'],name ='Recovered'),
    row=1, col=1
)


fig.add_trace(
    go.Bar(x=Top10CountriesConfirmed['Country'],y=Top10CountriesConfirmed['Count'],text = Top10CountriesConfirmed['Count'],name ='Confirmed'),
    row=2, col=1
)

fig.add_trace(
    go.Bar(x=Top10CountriesDeath['Country'],y=Top10CountriesDeath['Count'],text = Top10CountriesDeath['Count'],name ='Death'),
    row=3, col=1
)


fig.update_layout(height=1400, width=800, title_text="Top 10 Countries as of "+str(date))
fig.update_traces(texttemplate='%{text:.2s}', textposition='outside')
fig.show()
```

```python
def Calculatedays(df,NoofDays):
    dt= list(set(df['Date']))
    dt.sort()
    date = dt[-NoofDays:]
    return date
```

```python
last50days= Calculatedays(totaldataGhana, 50)
```

```python
total50daysGhana = totaldataGhana.loc[totaldataGhana['Date'].isin(last50days)]
```
